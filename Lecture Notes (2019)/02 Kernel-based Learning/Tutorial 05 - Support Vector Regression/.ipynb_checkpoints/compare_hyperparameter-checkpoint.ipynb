```python
%matplotlib inline

import numpy as np
from BA_SVR import svr_mj
import matplotlib.pyplot as plt
```

```python
#Generate sample data
X = np.sort(10*np.random.rand(100,1),axis=0)
y = np.sin(X).ravel()
y[::5] += 5*(0.5-np.random.rand(20))
```

# loss function

```python
'''
loss: 
['epsilon-insensitive','huber','laplacian','gaussian','polynomial','piecewise_polynomial'] 
defualt: epsilon-insensitive
p: polynomial/'piecewise_polynomial loss function p
sigma: for loss function
'''
svr_1 = svr_mj(loss='epsilon-insesitive',kernel='rbf',C=1.0,epsilon=0.1,gamma=0.1)
svr_1.fit(X,y)
pred_1 = svr_1.predict(X)
svr_2 = svr_mj(loss='huber',kernel='rbf',C=1.0,epsilon=0.1,gamma=0.1)
svr_2.fit(X,y)
pred_2 = svr_2.predict(X)
svr_3 = svr_mj(loss='laplacian',kernel='rbf',C=1.0,epsilon=0.1,gamma=0.1)
svr_3.fit(X,y)
pred_3 = svr_3.predict(X)
svr_4 = svr_mj(loss='gaussian',kernel='rbf',C=1.0,epsilon=0.1,gamma=0.1)
svr_4.fit(X,y)
pred_4 = svr_4.predict(X)
svr_5 = svr_mj(loss='polynomial',kernel='rbf',C=1.0,epsilon=0.1,gamma=0.1)
svr_5.fit(X,y)
pred_5 = svr_5.predict(X)
svr_6 = svr_mj(loss='piecewise_polynomial',kernel='rbf',C=1.0,epsilon=0.1,gamma=0.1)
svr_6.fit(X,y)
pred_6 = svr_6.predict(X)
```

```python
plt.scatter(X, y, color='darkorange', label='data')
plt.plot(X, pred_1, color='b', label='epsilon-insesitive')
plt.plot(X, pred_2, color='g', label='huber')
plt.plot(X, pred_3, color='r', label='laplacian')
plt.plot(X, pred_4, color='c', label='gaussian')
plt.plot(X, pred_5, color='m', label='polynomial')
plt.plot(X, pred_6, color='y', label='piecewise_polynomial')
plt.xlabel('X')
plt.ylabel('y')
plt.title('type of loss functions')
plt.legend()
plt.show()
```

# kernel

```python
svr_1 = svr_mj(loss='epsilon-insesitive',kernel='linear',epsilon=0.1)
svr_1.fit(X,y)
pred_1 = svr_1.predict(X)
svr_2 = svr_mj(loss='epsilon-insesitive',kernel='poly',epsilon=0.1)
svr_2.fit(X,y)
pred_2 = svr_2.predict(X)
svr_3 = svr_mj(loss='epsilon-insesitive',kernel='rbf',epsilon=0.1)
svr_3.fit(X,y)
pred_3 = svr_3.predict(X)
```

```python
plt.scatter(X, y, color='darkorange', label='data')
plt.plot(X, pred_1, color='b', label='linear')
plt.plot(X, pred_2, color='g', label='poly')
plt.plot(X, pred_3, color='r', label='rbf')
plt.xlabel('X')
plt.ylabel('y')
plt.title('type of kernel')
plt.legend()
plt.show()
```

# epsilon

```python
svr_1 = svr_mj(loss='epsilon-insesitive',kernel='poly',epsilon=0.01)
svr_1.fit(X,y)
pred_1 = svr_1.predict(X)
svr_2 = svr_mj(loss='epsilon-insesitive',kernel='poly',epsilon=0.1)
svr_2.fit(X,y)
pred_2 = svr_2.predict(X)
svr_3 = svr_mj(loss='epsilon-insesitive',kernel='poly',epsilon=1)
svr_3.fit(X,y)
pred_3 = svr_3.predict(X)
```

```python
plt.scatter(X, y, color='darkorange', label='data')
plt.plot(X, pred_1, color='b', label='epsilon 0.01')
plt.plot(X, pred_2, color='g', label='epsilon 0.1')
plt.plot(X, pred_3, color='r', label='epsilon 1')
plt.xlabel('X')
plt.ylabel('y')
plt.title('type of loss functions')
plt.legend()
plt.show()
```

# C

```python
svr_1 = svr_mj(loss='epsilon-insesitive',kernel='rbf',C=1,epsilon=1,gamma=0.1)
svr_1.fit(X,y)
pred_1 = svr_1.predict(X)
svr_2 = svr_mj(loss='epsilon-insesitive',kernel='rbf',C=10,epsilon=1,gamma=0.1)
svr_2.fit(X,y)
pred_2 = svr_2.predict(X)
svr_3 = svr_mj(loss='epsilon-insesitive',kernel='rbf',C=100,epsilon=1,gamma=0.1)
svr_3.fit(X,y)
pred_3 = svr_3.predict(X)
svr_4 = svr_mj(loss='epsilon-insesitive',kernel='rbf',C=1000,epsilon=1,gamma=0.1)
svr_4.fit(X,y)
```

```python
plt.scatter(X, y, color='darkorange', label='data')
plt.plot(X, pred_1, color='b', label='C 1')
plt.plot(X, pred_2, color='g', label='C 10')
plt.plot(X, pred_3, color='r', label='C 100')
plt.plot(X, pred_3, color='c', label='C 1000')
plt.xlabel('X')
plt.ylabel('y')
plt.title('C')
plt.legend()
plt.show()
```
