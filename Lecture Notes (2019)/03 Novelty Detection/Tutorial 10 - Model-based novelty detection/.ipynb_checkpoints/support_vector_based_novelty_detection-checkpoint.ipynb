# Setup

```python
# 필요 라이브러리
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets
from sklearn.preprocessing import StandardScaler
from scipy.optimize import minimize

iris = datasets.load_iris()
scaler = StandardScaler()

raw_X = iris["data"][:, [2, 3]]  # petal length, petal width
raw_y = iris["target"]
```

# EDA

```python
plt.figure(figsize=(10,2.7))
plt.plot(raw_X[:, 0][raw_y==2], raw_X[:, 1][raw_y==2], "g^", label="Virginica")
plt.plot(raw_X[:, 0][raw_y==1], raw_X[:, 1][raw_y==1], "bs", label="Versicolor")
plt.plot(raw_X[:, 0][raw_y==0], raw_X[:, 1][raw_y==0], "yo", label="Setosa")
plt.legend(loc="upper left", fontsize=14)
plt.xlabel("Petal length", fontsize=14)
plt.ylabel("Petal width", fontsize=14)

plt.show()
```

# Modules

```python
def plot_boundary(svm, scope, X, y, *args):
    svm(scope, X, y, args)
    if len(y[y==2]):
        plt.plot(X[:, 0][y==2], X[:, 1][y==2], "g^", label="Virginica")
    if len(y[y==1]):
        plt.plot(X[:, 0][y==1], X[:, 1][y==1], "bs", label="Versicolor")
    if len(y[y==0]):
        plt.plot(X[:, 0][y==0], X[:, 1][y==0], "yo", label="Setosa")
    plt.xlabel("Petal length", fontsize=14)
    plt.ylabel("Petal width", fontsize=14)
    plt.legend(loc="upper left", fontsize=14)
    plt.axis(scope)
```

# Sensitivity to feature scales

```python
def hard_margin(scope, X, y, args):
    xmin, xmax, ymin, ymax = scope
    n = len(y)
    y_ = y.copy()
    y_[[y==0]] = -1
    alpha = [0] * n
    def target(alpha):
        w = (alpha * y_).dot(X)
        return 1 / 2 * w.dot(w.T) - sum(alpha)
    cons = ({'type': 'eq', 'fun': lambda alpha: alpha.dot(y_)})
    bnds = [[0, None]] * n
    res = minimize(target, alpha, options={'disp': True}, bounds=bnds, constraints=cons)
    alpha_ = np.float16(res.x)
    w = (alpha_ * y_).dot(X)
    SVI = np.where(alpha_ != 0)[0]
    svs = X[SVI]
    
    plt.scatter(svs[:, 0], svs[:, 1], s=180, facecolors='#FFAAAA')
        
    all_b = y_[SVI] - w.T.dot(X[SVI].T)
    b = all_b[0]
    
    x0 = np.linspace(xmin, xmax, 200)
    decision_boundary = -w[0]/w[1] * x0 - b/w[1]

    margin = 1/w[1]
    gutter_up = decision_boundary + margin
    gutter_down = decision_boundary - margin

    plt.scatter(svs[:, 0], svs[:, 1], s=180, facecolors='#FFAAAA')
    plt.plot(x0, decision_boundary, "k-", linewidth=2)
    plt.plot(x0, gutter_up, "k--", linewidth=2)
    plt.plot(x0, gutter_down, "k--", linewidth=2)
```

> From
> ## w0 x0 + w1 x1 + b = 0
> To
> ## x1 = -w0 / w1 * x0 - b / w1


```python
# 데이타 정의
setosa_or_versicolor = (raw_y == 0) | (raw_y == 1)
X = raw_X[setosa_or_versicolor]
scaled_X = scaler.fit_transform(X)
y = raw_y[setosa_or_versicolor]

# 스케일링 하지 않고 알고리즘 적용
plt.figure(figsize=(12,2.7))
plot_boundary(hard_margin, [0, 5.5, 0, 2], X, y)
plt.show()

# 스케일링 한 후 알고리즘 적용
plt.figure(figsize=(12,2.7))
plot_boundary(hard_margin, [-2, 2, -2, 2], scaled_X, y)
plt.show()
```

# Sensitivity to outliers

```python
# 데이타에 아웃라이어를 삽입
X_outliers = np.array([[3.2, 0.8]])
y_outliers = np.array([0])
Xo = np.concatenate([X, X_outliers], axis=0)
yo = np.concatenate([y, y_outliers], axis=0)

# 1행 그래프
plt.figure(figsize=(12,2.7))
plt.plot(raw_X[:, 0][raw_y==2], raw_X[:, 1][raw_y==2], "g^", label="Virginica")
plt.plot(raw_X[:, 0][raw_y==1], raw_X[:, 1][raw_y==1], "bs", label="Versicolor")
plt.plot(raw_X[:, 0][raw_y==0], raw_X[:, 1][raw_y==0], "yo", label="Setosa")
plt.legend(loc="upper left", fontsize=14)
plt.xlabel("Petal length", fontsize=14)
plt.ylabel("Petal width", fontsize=14)
plt.show()

plt.figure(figsize=(12,2.7))

# 2행 1열 그래프
plt.subplot(121)
plot_boundary(hard_margin, [0, 5.5, 0, 2], Xo, yo)
plt.annotate("Outlier",
             xy=(X_outliers[0][0], X_outliers[0][1]),
             xytext=(2.5, 1.7),
             ha="center",
             arrowprops=dict(facecolor='black', shrink=0.1),
             fontsize=16,
            )

# 2행 2열 그래프
plt.subplot(122)
plt.plot(raw_X[:, 0][raw_y==2], raw_X[:, 1][raw_y==2], "g^")
plt.plot(raw_X[:, 0][raw_y==1], raw_X[:, 1][raw_y==1], "bs")
plt.text(3, 2, "Impossible!", fontsize=24, color="red")
plt.xlabel("Petal length", fontsize=14)
plt.ylabel("Petal width", fontsize=14)

plt.show()
```

# Large margin *vs* margin violations

```python
def soft_margin(scope, X, y, args):
    xmin, xmax, ymin, ymax = scope
    C, = args
    n = len(X)
    y_ = y.copy()
    y_[[y==2]] = -1
    alpha = [0] * n
    def target(alpha):
        w = (alpha * y_).dot(X)
        return 1 / 2 * w.dot(w.T) - sum(alpha)
    cons = ({'type': 'eq', 'fun': lambda alpha: alpha.dot(y_)})
    bnds = [[0, C]] * n
    res = minimize(target, alpha, bounds=bnds, constraints=cons)
    alpha_ = np.float16(res.x)
    w = (alpha_ * y_).dot(X)
    SVI = np.where(alpha_ != 0)[0]
    svs = X[SVI]
    MSVI = SVI[alpha_[SVI] < C]
    msvs = X[MSVI]
        
    plt.scatter(svs[:, 0], svs[:, 1], s=180, facecolors='#FFAAAA')
    plt.scatter(msvs[:, 0], msvs[:, 1], s=180, facecolors='#AAAAAA')
    
    all_b = y_[MSVI] - w.T.dot(X[MSVI].T)
    b = all_b[0]
    plt.title("C = {}".format(C), fontsize=16)
    
    x0 = np.linspace(xmin, xmax, 200)
    decision_boundary = -w[0]/w[1] * x0 - b/w[1]
    margin = 1/w[1]
    gutter_up = decision_boundary + margin
    gutter_down = decision_boundary - margin
    plt.plot(x0, decision_boundary, "k-", linewidth=2)
    plt.plot(x0, gutter_up, "k--", linewidth=2)
    plt.plot(x0, gutter_down, "k--", linewidth=2)
```

```python
virginica_or_versicolor = (raw_y == 1) | (raw_y == 2)
X = scaler.fit_transform(raw_X[virginica_or_versicolor])
y = raw_y[virginica_or_versicolor]

for i in range(3):
    C = 10 ** i
    plt.figure(figsize=(12,2.7))
    plot_boundary(soft_margin, [-2, 2, -2, 2], X, y, C)
    plt.show()
```

# Non-linear classification

```python
def linear_kernel(X1, X2):
    return X1.dot(X2.T)

def polynomial_kernel(X1, X2):    
    return (1 + X1.dot(X2.T)) ** 2

def gaussian_kernel(X1, X2):
    return np.exp(-(X1 * X1)).dot(np.exp(-(X2 * X2)).T) * np.exp(2 * X1.dot(X2.T))
```

```python
def kernel_tricked(scope, X, y, args):
    xmin, xmax, ymin, ymax = scope
    C, kernel = args
    n = len(X)
    y_ = y.copy()
    y_[[y==2]] = -1
    alpha = [0] * n
    
    kX = kernel(X, X)
    
    def target(alpha):
        return 1 / 2 * (alpha * y_).dot(kX).dot((alpha * y_).T) - sum(alpha)
    
    cons = ({'type': 'eq', 'fun': lambda alpha: alpha.dot(y_)})
    bnds = [[0, C]] * n
    res = minimize(target, alpha, bounds=bnds, constraints=cons)
    
    alpha_ = res.x
    SVI = np.where(np.float16(alpha_) != 0)[0]
    svs = X[SVI]
    MSVI = SVI[alpha_[SVI] < C - 0.001]
    msvs = X[MSVI]
        
    plt.scatter(svs[:, 0], svs[:, 1], s=180, facecolors='#FFAAAA')
    plt.scatter(msvs[:, 0], msvs[:, 1], s=180, facecolors='#AAAAAA')
    
    x0s = np.linspace(xmin, xmax, 200)
    x1s = np.linspace(ymin, ymax, 200)
    
    x0, x1 = np.meshgrid(x0s, x1s)
    X_points = np.c_[x0.ravel(), x1.ravel()]
    plt.title("C = {}, {}".format(C, kernel.__name__), fontsize=16)
    
    def prediction(X_points):
        bs = y_[MSVI] - (alpha_ * y_).dot(kernel(X, msvs))
        b = bs[0]
        result = (alpha_ * y_).dot(kernel(X, X_points)) + b
        result[result >= 0] = 1
        result[result < 0] = -1
        return result
    
    y_pred = prediction(X_points).reshape(x0.shape)
    plt.contourf(x0, x1, y_pred, cmap=plt.cm.brg, alpha=0.2)
```

```python
virginica_or_versicolor = (raw_y == 1) | (raw_y == 2)
X = scaler.fit_transform(raw_X[virginica_or_versicolor])
y = raw_y[virginica_or_versicolor]

for kernel in [linear_kernel, polynomial_kernel, gaussian_kernel]:
    for C in [1, 100]:
        plt.figure(figsize=(12,2.7))
        plot_boundary(kernel_tricked, [-2, 2, -2, 2], X, y, C, kernel)
        plt.show()
```

# One-SVM novelty detection

```python
def one_svm(scope, X, y, args):
    xmin, xmax, ymin, ymax = scope
    V, kernel = args
    l = len(X)
    C = 1 / (l * V)
    alpha = [0] * l
    kX = kernel(X, X)
    
    cons = ({'type': 'eq', 'fun': lambda alpha: np.sum(alpha) - 1})
    bnds = [[0, C]] * l
    
    def target(alpha):
        return 1 / 2 * (alpha).dot(kX).dot((alpha).T)
    
    res = minimize(target, alpha, bounds=bnds, constraints=cons)
    
    alpha_ = np.float16(res.x)
    SVI, = np.where(alpha_ != 0)
    svs = X[SVI]
    MSVI = SVI[alpha_[SVI] < C]
    msvs = X[MSVI]
        
    plt.scatter(svs[:, 0], svs[:, 1], s=180, facecolors='#FFAAAA')
    plt.scatter(msvs[:, 0], msvs[:, 1], s=180, facecolors='#AAAAAA')
    plt.title("V = {}, {}".format(V, kernel.__name__), fontsize=16)
    
    x0s = np.linspace(xmin, xmax, 200)
    x1s = np.linspace(ymin, ymax, 200)
    
    x0, x1 = np.meshgrid(x0s, x1s)
    X_points = np.c_[x0.ravel(), x1.ravel()]

    def prediction(X_points):
        rhos = (alpha_).dot(kernel(X, msvs))
        print(rhos)
        rho = rhos[0]
        result = (alpha_).dot(kernel(X, X_points)) - rho
        result[result >= 0] = 1
        result[result < 0] = -1
        return result
    
    y_pred = prediction(X_points).reshape(x0.shape)
    plt.contourf(x0, x1, y_pred, cmap=plt.cm.brg, alpha=0.2)
```

```python
virginica = (raw_y == 2)
y = raw_y[virginica]

for kernel in [linear_kernel, polynomial_kernel, gaussian_kernel]:
    for V in [0.1, 0.05]:
        plt.figure(figsize=(12,2.7))
        try:
            X = raw_X[virginica]
            plot_boundary(one_svm, [4, 8, 0, 5], X, y, V, kernel)
        except:
            scaled_X = scaler.fit_transform(raw_X[virginica])
            plot_boundary(one_svm, [-3, 3, -3, 3], scaled_X, y, V, kernel)
        plt.show()
```

# SVDD novelty detection

```python
def svdd(scope, X, y, args):
    xmin, xmax, ymin, ymax = scope
    V, kernel = args
    l = len(X)
    C = 1 / (l * V)
    alpha = [0] * l    
    kX = kernel(X, X)
    
    cons = ({'type': 'eq', 'fun': lambda alpha: sum(alpha) - 1})
    bnds = [[0, C]] * l
    
    def target(alpha):
        return alpha.dot(kX).dot(alpha.T) - alpha.dot(np.diag(kX))

    res = minimize(target, alpha, bounds=bnds, constraints=cons)
    
    alpha_ = np.float16(res.x)
    SVI = np.where(alpha_ != 0)[0]
    svs = X[SVI]
    MSVI = SVI[alpha_[SVI] < C]
    msvs = X[MSVI]
        
    plt.scatter(svs[:, 0], svs[:, 1], s=180, facecolors='#FFAAAA')
    plt.scatter(msvs[:, 0], msvs[:, 1], s=180, facecolors='#AAAAAA')
    plt.title("V = {}, {}".format(V, kernel.__name__), fontsize=16)
    
    x0s = np.linspace(xmin, xmax, 50)
    x1s = np.linspace(ymin, ymax, 50)
    
    x0, x1 = np.meshgrid(x0s, x1s)
    X_points = np.c_[x0.ravel(), x1.ravel()]

    def prediction(X_points):
        
        rss = (np.diag(kernel(msvs, msvs)) +
               alpha_.dot(kX).dot(alpha_.T) - 2 * alpha_.dot(kernel(X, msvs)))
        print(rss)
        rs = rss[0]
        result = rs - (np.diag(kernel(X_points, X_points)) +
                       alpha_.dot(kX).dot(alpha_.T) -
                       2 * alpha_.dot(kernel(X, X_points)))
        result[result >= 0] = 1
        result[result < 0] = -1
        return result
    
    y_pred = prediction(X_points).reshape(x0.shape)
    plt.contourf(x0, x1, y_pred, cmap=plt.cm.brg, alpha=0.2)
```

```python
virginica = (raw_y == 2)
scaled_X = scaler.fit_transform(raw_X[virginica])
y = raw_y[virginica]

for kernel in [linear_kernel, polynomial_kernel, gaussian_kernel]:
    plt.figure(figsize=(12,2.7))
    plot_boundary(svdd, [-3, 3, -3, 3], scaled_X, y, 0.05, kernel)
    plt.show()
```
